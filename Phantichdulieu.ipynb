```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Đọc 1000 dòng đầu tiên của dữ liệu (thay thế 'your_data.csv' bằng tên file của bạn)
data = pd.read_csv('Churn_Modelling.csv', nrows= 1000)
data = data.dropna(axis='columns') # drop columns with NaN
data = data[[col for col in data if data[col].nunique() > 1]] # keep columns where there are more than 1 unique values

# Lọc ra các khách hàng đã rời bỏ
customers_left = data[data['Exited'] == 1]

# Đếm số lượng khách hàng nam và nữ đã rời bỏ
gender_count = customers_left['Gender'].value_counts()
print(gender_count)

# Vẽ biểu đồ cột
gender_count.plot(kind='bar')
plt.title('Số lượng khách hàng nam và nữ đã rời bỏ')
plt.xlabel('Giới tính')
plt.ylabel('Số lượng khách hàng')
plt.show()
```

```python

# Lọc ra các khách hàng có Tenure từ 8 đến 10 năm
customers_tenure_8_10 = data[(data['Tenure'] >= 8) & (data['Tenure'] <= 10)]

# Tính số lượng khách hàng ở lại và rời bỏ trong nhóm trên
customer_count = customers_tenure_8_10['Exited'].value_counts()

# Vẽ biểu đồ cột
customer_count.plot(kind='bar')
plt.title('Số lượng khách hàng ở lại và rời bỏ (8-10 năm)')
plt.xlabel('Exited')
plt.ylabel('Số lượng khách hàng')
plt.xticks([0, 1], ['Ở lại', 'Rời bỏ'])
plt.show()
```

```python
# Lọc ra các khách hàng có Tenure từ 2 đến 8 năm
customers_tenure_8_10 = data[(data['Tenure'] >= 2) & (data['Tenure'] < 8)]

# Tính số lượng khách hàng ở lại và rời bỏ trong nhóm trên
customer_count = customers_tenure_8_10['Exited'].value_counts()

# Vẽ biểu đồ cột
customer_count.plot(kind='bar')
plt.title('Số lượng khách hàng ở lại và rời bỏ (2 - 7 năm)')
plt.xlabel('Exited')
plt.ylabel('Số lượng khách hàng')
plt.xticks([0, 1], ['Ở lại', 'Rời bỏ'])
plt.show()
```

```python
# Lọc ra các khách hàng có Tenure từ 2 đến 8 năm
customers_tenure_8_10 = data[(data['Tenure'] < 2)]

# Tính số lượng khách hàng ở lại và rời bỏ trong nhóm trên
customer_count = customers_tenure_8_10['Exited'].value_counts()

# Vẽ biểu đồ cột
customer_count.plot(kind='bar')
plt.title('Số lượng khách hàng ở lại và rời bỏ (dưới 2 năm)')
plt.xlabel('Exited')
plt.ylabel('Số lượng khách hàng')
plt.xticks([0, 1], ['Ở lại', 'Rời bỏ'])
plt.show()
```

```python

# customers_tenure_8_10 = data[(data['Tenure'] >= 8) & (data['Tenure'] <= 10)]
# Vẽ biểu đồ countplot để so sánh số lượng khách hàng rời bỏ và ở lại theo từng nhóm Tenure
sns.countplot(x='Tenure', hue='Exited', data=data)

# Đặt tiêu đề và nhãn cho biểu đồ
plt.title('Số lượng khách hàng rời bỏ và ở lại theo Tenure')
plt.xlabel('Thời gian sử dụng dịch vụ (Tenure)')
plt.ylabel('Số lượng khách hàng')
plt.legend(title='Exited', labels=['Ở lại', 'Rời bỏ'])

# Hiển thị biểu đồ
plt.show()
```
